```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
name=[]
n=[]
for i in soup.find('td', class_="rankings-block__top-player-container"):
    name.append(i.text.replace("\n","").split(",")[0])
n=name[:2]

for j in soup.find_all('td', class_="table-body__cell rankings-table__name name"):
    n.append(j.text.replace('\n',''))
Name=n[1:11]
len(Name)
Name
```

```python
rating=[] 
rate=[]
for i in soup.find('td' ,class_="u-text-left"):
    rating.append(i.text.replace('\n',"").split(',')[0])
rate=rating[:2]
for i in soup.find_all('td' ,class_="table-body__cell rating"):
    rate.append(i.text)
r=rate[1:11]
r
```

```python
team=[]
for i in soup.find_all('div' ,class_="rankings-block__banner--nationality"):
    team.append(i.text.replace('\n',''))
for i in soup.find_all('span', class_="table-body__logo-text"):
    team.append(i.text.replace('\n',''))
t=team[0:10]
t
```
