#imdb top 50 indian movies

```python
from bs4 import BeautifulSoup
import requests

page=requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating')

page

soup=BeautifulSoup(page.content)
soup



rating=[]
for i in soup.find_all('div', class_="inline-block ratings-imdb-rating" ):
    rating.append(i.text.replace('\n',""))
rating

name=[]
for i in soup.find_all('h3' ,class_="lister-item-header"):
    name.append(i.text.replace('\n',""))
name

year=[]
for i in soup.find_all('span', class_="lister-item-year text-muted unbold"):
    year.append(i.text)
year

print(len(name),len(year),len(rating))

import pandas as pd
df=pd.DataFrame({'Movie_Name':name,"Year_of_Release":year,'Rating':rating})

df

```

```python
from bs4 import BeautifulSoup
import requests

page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page

soup=BeautifulSoup(page.content)
soup

filmname=[]
for i in soup.find_all('td', class_="titleColumn"):
    filmname.append(i.text.replace('\n',""))
n=filmname[0:50]
n

year=[]
for i in soup.find_all('span', class_="secondaryInfo"):
    year.append(i.text)
y=year[0:50] 
y

rating=[]
for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',""))
r=rating[0:50]
r

print (len(n),len(y),len(r))

import pandas as pd
df=pd.DataFrame({'Movie_Name':n,'Year_of_Release':y,'Rating':r})
df

```

```python
from bs4 import BeautifulSoup
import requests

page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page

soup=BeautifulSoup(page.content)
soup

team=[]
for i in soup.find_all('span', class_="u-hide-phablet"):
    team.append(i.text.split(',')[0])
t=team[1:11]
t

matches=[]
for i in soup.find('td',class_="rankings-block__banner--matches"):
    matches.append(i.text.split(",")[0])
for j in soup.find_all('td', class_="table-body__cell u-center-text"):
    matches.append(j.text.split(",")[0])
m=matches[0:10]
m


points=[]
for p in soup.find_all('td', class_="rankings-block__banner--points"):
    points.append(p.text)
for i in soup.find_all('td' ,class_="table-body__cell u-center-text"):
    points.append(i.text)
p=points[::2]
po=p[0:10]
po

rating=[]
for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    rating.append(i.text.replace('\n',""))
for j in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    rating.append(j.text)
r=rating[0:10]
r

print(len(t),len(m),len(po),len(r))

import pandas as pd
df=pd.DataFrame({'Team_Name':t,"Matches":m,"Points":po,"Rating":r})
df

```
